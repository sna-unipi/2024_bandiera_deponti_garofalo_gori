# Graphs for Edge Removal

In this notebook we plot some graphs about edge removal and network dismantle.

```python
# Import libreries.
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations, product
import random
```

```python
# Import arrays.
import pickle
dir = "c:/Users/HP/Desktop/UNI/LM_1/SNA/SNA/data/"

with open(f'{dir}removed_edges_overlap_decreasing.pkl', 'rb') as file:
    re_overlap_decreasing = pickle.load(file)

with open(f'{dir}components_overlap_decreasing.pkl', 'rb') as file:
    comps_overlap_decreasing = pickle.load(file)

with open(f'{dir}removed_edges_overlap_increasing.pkl', 'rb') as file:
    re_overlap_increasing = pickle.load(file)

with open(f'{dir}components_overlap_increasing.pkl', 'rb') as file:
    comps_overlap_increasing = pickle.load(file)

with open(f'{dir}removed_edges_random.pkl', 'rb') as file:
    re_random = pickle.load(file)

with open(f'{dir}components_random.pkl', 'rb') as file:
    comps_random = pickle.load(file)

with open(f'{dir}removed_edges_weight_increasing.pkl', 'rb') as file:
    re_weight_increasing = pickle.load(file)

with open(f'{dir}components_weight_increasing.pkl', 'rb') as file:
    comps_weight_increasing = pickle.load(file)

with open(f'{dir}removed_edges_weight_decreasing.pkl', 'rb') as file:
    re_weight_decreasing = pickle.load(file)

with open(f'{dir}components_weight_decreasing.pkl', 'rb') as file:
    comps_weight_decreasing = pickle.load(file)

with open(f'{dir}removed_edges_duration_increasing.pkl', 'rb') as file:
    re_duration_increasing = pickle.load(file)

with open(f'{dir}components_duration_increasing.pkl', 'rb') as file:
    comps_duration_increasing = pickle.load(file)

with open(f'{dir}removed_edges_duration_decreasing.pkl', 'rb') as file:
    re_duration_decreasing = pickle.load(file)

with open(f'{dir}components_duration_decreasing.pkl', 'rb') as file:
    comps_duration_decreasing = pickle.load(file)
```

```python
# Plot graphs.
plt.figure(figsize=(15,8))
plt.plot(re_random, comps_random, label='random')
plt.plot(re_overlap_decreasing, comps_overlap_decreasing, label='decreasing overlap', color='green')
plt.plot(re_overlap_increasing, comps_overlap_increasing, label='increasing overlap', color='green', linestyle='-.')
plt.plot(re_weight_decreasing, comps_weight_decreasing, label='decreasing weight', color='red')
plt.plot(re_weight_increasing, comps_weight_increasing, label='increasing weight', color='red', linestyle='-.')
# plt.plot(re_duration_decreasing, comps_duration_decreasing, label='decreasing duration', color='orange', alpha=0.5)
# plt.plot(re_duration_increasing, comps_duration_increasing, label='increasing duration', color='orange', linestyle='-.')
plt.xlabel("Removed Edges",size=18)
plt.ylabel("Giant Component size",size=18)
plt.title('Different removal strategies', size=20)
plt.legend(fontsize=15)
plt.show()
```
